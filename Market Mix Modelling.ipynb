```python
data1=pd.read_csv('Secondfile.csv')
data1.info
```

```python
data1=data1[['total_gmv','TV','Digital',"Sponsorship",'Content.Marketing','Online.marketing','Affiliates',"SEM",'Radio','Other']]
data1.info
```

```python
sns.pairplot(data = data1,
            x_vars = ['TV','Digital',"Sponsorship",'Content.Marketing','Online.marketing','Affiliates',"SEM",'Radio','Other'],
            y_vars = 'total_gmv',
            size = 7,
            kind = 'reg')
```

```python
data1.info
data1=data1.fillna(0)
data1.info
```

```python
x = data1.iloc[:,1:10]
y = data1.iloc[:,0]
x
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression

X_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 0)
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
y_prediction = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score as r2, mean_squared_error as mse
r2(y_test, y_prediction)
```

```python
mse(y_test, y_prediction)
```
